```python
import pandas as pd
import pyproj
import folium
```

```python
path_2 = "/work/markers/UO-markers.csv"
df = pd.read_csv(path_2, delimiter=";")
df['X'] = df['X'].str.replace(',', '.').astype(float)
df['Y'] = df['Y'].str.replace(',', '.').astype(float)
df
```

```python
# path_2 = "/work/wellBH.csv"
# df = pd.read_csv(path_2, delimiter=";")
df['longitude'] = 1.1111
df['latitude'] = 1.1111
# Define the EPSG codes for the source and target CRS
src_crs = pyproj.CRS("EPSG:3176")
ll_crs = pyproj.CRS("EPSG:4326")
#Function of converting
def transform(src_crs, ll_crs, point):
    # Create a transformer object to perform the transformation
    transformer = pyproj.Transformer.from_crs(src_crs, ll_crs)
    # Use the transformer to perform the transformation
    lon, lat = transformer.transform(*point)
    # Print the new coordinates (lat, long in EPSG 4326)
    print(f"New coordinates: ({lon:.6f}, {lat:.6f})")
    return round(lon, 7),round(lat, 7)
#Add field of lat/long after the transformer
for i in range(len(df)):
    lon, lat = transform(src_crs, ll_crs, (df.X[i], df.Y[i]))
    df['longitude'][i]=lon
    df['latitude'][i]=lat
df.to_csv('/work/markers/converted/UO_markers_4326.csv')
```

```python
path = '/work/VSP_data.xlsx - coordinate.csv'
df = pd.read_csv(path, usecols=['well_id','x', 'y', 'd1', 'd2']).rename(columns={'well_id':'WELL_ID','x':'X', 'y':'Y','d1':'TVD', 'd2':'MD'})
df
```

```python
df['longitude'] = 1.1111
df['latitude'] = 1.1111
# Define the EPSG codes for the source and target CRS
src_crs = pyproj.CRS("EPSG:3176")
ll_crs = pyproj.CRS("EPSG:4326")
#Function of converting
def transform(src_crs, ll_crs, point):
    # Create a transformer object to perform the transformation
    transformer = pyproj.Transformer.from_crs(src_crs, ll_crs)
    # Use the transformer to perform the transformation
    lon, lat = transformer.transform(*point)
    # Print the new coordinates (lat, long in EPSG 4326)
    print(f"New coordinates: ({lon:.6f}, {lat:.6f})")
    return round(lon, 7),round(lat, 7)
#Add field of lat/long after the transformer
for i in range(len(df)):
    lon, lat = transform(src_crs, ll_crs, (df.X[i], df.Y[i]))
    df['longitude'][i]=lon
    df['latitude'][i]=lat
```

```python
df.to_csv('/work/VSP_lat_long.csv')
df
```

```python
# Create a Folium map centered on the first point in the DataFrame
map_center = [df['latitude'].iloc[0], df['longitude'].iloc[0]]
my_map = folium.Map(location=map_center, zoom_start=12)

# Add each point in the DataFrame as a marker to the map
for index, row in df.iterrows():
    folium.Marker(location=[row['longitude'], row['latitude']], 
                  popup=row['WELL_ID']).add_to(my_map)
# Display the map
my_map
```
